### Importing the dataset

```python
# Importing pandas library
import pandas as pd
import numpy as np
# Importing the csv data files 
data = pd.read_csv('auto-mpg.csv',error_bad_lines=False, warn_bad_lines=True)

# Preinting some information about the dataset
data.head()
```

```python
print (data.shape)
print ('Columns of dataset- \n', data.columns)
data_temp = data

print (data.info())
```

### Data Cleaning

```python
# Function which counts the number of null values for each column
def count_null_values(dataset, column_list):
    for i in range (len(column_list)):
        print ("The total number of null values in :",column_list[i])
        print (dataset[column_list[i]].isnull().sum())
    return

def print_uniques(dataset, column_list):
    for i in range (len(column_list)):
        print ("Unique values for the column:",column_list[i])
        print (dataset[column_list[i]].unique())
        print ('\n')
    return

# Printing the null and unique values for each attribute in the dataset
print_uniques(data, data.columns)
count_null_values(data, data.columns)
```

```python
# Converting the horsepower from object to integer
data_clean = data
# Changing the missing values from '?' to '0'
data_clean['horsepower']  = data_clean['horsepower'] .replace('?', '0')
data_clean['horsepower'] = data_clean['horsepower'].astype(str).astype(int)
data_clean['horsepower']  = data_clean['horsepower'] .replace(0, np.nan)
print (data_clean.info())
```

```python
print_uniques(data_clean, data_clean.columns)
count_null_values(data_clean, data_clean.columns)
```

```python
data_clean.describe()
```

```python
# Dealing with the missing horsepower values with the median 
data_clean['horsepower']= data_clean['horsepower'].replace(np.nan, 93.5)
count_null_values(data_clean, data_clean.columns)
```

### Exploratory Data Analysis

```python
# Finding out the correlation between the features

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Storing the converted data
df_hash = pd.DataFrame()
df_hash['car name'] = data_clean['car name']

sns.set(style="white")

temp = data_clean.apply(LabelEncoder().fit_transform)
data_label_encoding = data_clean
data_label_encoding['car name'] = temp['car name']  

df_hash['encoded'] = temp['car name'] 
# Compute the correlation matrix
corr = data_label_encoding.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 8))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.set_context("paper", rc={"font.size":12,"axes.titlesize":12, "axes.labelsize":12}) 
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
ax.set_title('Correlation Between The Pairs of Columns ')
plt.show()
```

```python
# Distribution of mpg in the dataset
from scipy import stats
sns.distplot(data_clean["mpg"], kde=False, fit=stats.gamma).set(title = 'Distribution for mpg in the dataset', xlabel = 'mpg', ylabel = 'Proportion Distribution' )
plt.show()
```

Preparing the final dataset

```python
final_df = data_label_encoding
final_df['car name'] = data['car name']
print(final_df.head())
```

Splitting the dataset for machine learning

```python
from sklearn.model_selection import train_test_split

# Defining the X and y 
X = final_df.drop('mpg', axis=1).values
y = final_df['mpg'].values

# Partitioning the dataset
X_train,X_test,y_train,y_test=train_test_split(X, y,test_size=0.3,random_state=42)
print(X_train.shape, X_test.shape)

# Conducting PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_PCA = pca.fit_transform(X)

X_PCA_train, X_PCA_test, y_PCA_train, y_PCA_test = train_test_split(X_PCA, y, test_size= 0.3, random_state=42)
print ('PCA Dataset')
print (X_PCA_train.shape)
print (X_PCA_test.shape)
```

Creating the Machine learning model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# RandomForestRegressor model with hyper tuning
regr = RandomForestRegressor()
parameters = {'n_estimators': [10, 12, 15, 18], 'max_depth': [6, 7, 8, 9],  \
              'min_samples_split': [15, 20, 25], 'max_leaf_nodes': [15, 20, 25]}

randfor_regr = GridSearchCV(regr , parameters,  cv = 5)
randfor_regr.fit(X_train, y_train)
print(randfor_regr.best_params_)
print('Training accuracy: ',randfor_regr.best_score_)
```

```python

# Training the model with the entire training dataset
model = RandomForestRegressor(random_state=0, max_depth= 9, max_leaf_nodes= 25,\
                                       min_samples_split= 15, n_estimators= 10)
model.fit(X_train, y_train)
# Predicting the output with the test dataset
y_pred = model.predict(X_test)
# Getting the accuracy metric
sns.distplot(y_test - y_pred)
plt.show()
plt.scatter(y_test, y_pred)
plt.show()
```

```python
# Getting the accuracy on the test dataset
print(model.score(X_test,y_test))
```
